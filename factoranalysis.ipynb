“요인분석”이라고 하면 사실 서로 다른 3계열이 섞여 있어:

공통요인분석 (Common Factor Analysis)

특이값 분해 기반 (SVD / PCA)

음수 미포함 행렬분해 (NMF)

1️⃣ 공통요인분석 (Common Factor Analysis)
🔥 개념

X=ΛF+ϵ

Λ: 요인 적재량

F: 공통요인

ϵ: 고유요인(오차)

| 옵션                 | 의미     |
| ------------------ | ------ |
| rotation='varimax' | 직교 회전  |
| rotation='promax'  | 사각 회전  |
| method='ml'        | 최대우도   |
| method='principal' | 주성분 방식 |


```python
from factor_analyzer import FactorAnalyzer
from sklearn.datasets import load_iris
import pandas as pd

data = load_iris()
df = pd.DataFrame(data.data)

fa = FactorAnalyzer(n_factors=2, rotation='varimax')
fa.fit(df)

# 요인 적재량
loadings = fa.loadings_
print(loadings)

# 요인 점수
factor_scores = fa.transform(df)
```

2️⃣ SVD 기반 (PCA)
🔥 개념

X=UΣV

✔ 분산 최대 방향 찾음
✔ 잠재변수 모델 아님
✔ 차원축소용

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

X_scaled = StandardScaler().fit_transform(df)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(pca.components_)   # 주성분
print(pca.explained_variance_ratio_)

import numpy as np

U, S, Vt = np.linalg.svd(X_scaled, full_matrices=False)
```

| PCA      | 공통요인    |
| -------- | ------- |
| 전체 분산 설명 | 공통분산만   |
| 수학적 분해   | 통계적 모형  |
| 차원축소     | 잠재요인 해석 |


NMF (Non-negative Matrix Factorization)
🔥 개념

X≈WH

✔ 모든 값 ≥ 0
✔ 해석성 좋음
✔ 토픽모델링, 이미지 분해

```python
from sklearn.decomposition import NMF

# 음수 없어야 함
X_nonneg = df.values

nmf = NMF(n_components=2, init='random', random_state=42)
W = nmf.fit_transform(X_nonneg)
H = nmf.components_

print(W.shape)
print(H.shape)
```

| 특징          | 설명          |
| ----------- | ----------- |
| 음수 불가       | 데이터 >= 0 필요 |
| additive 구조 | 부분의 합       |
| 해석 용이       | 각 요인이 부분 구조 |

| 구분    | 공통요인 | PCA  | NMF |
| ----- | ---- | ---- | --- |
| 이론 기반 | 통계모형 | 선형대수 | 최적화 |
| 오차모형  | 있음   | 없음   | 없음  |
| 음수 허용 | 허용   | 허용   | 불가  |
| 해석성   | 높음   | 낮음   | 높음  |

| 목적      | 추천   |
| ------- | ---- |
| 설문 잠재요인 | 공통요인 |
| 차원 축소   | PCA  |
| 토픽 모델링  | NMF  |
| 노이즈 제거  | PCA  |
| 구조 해석   | 요인분석 |
